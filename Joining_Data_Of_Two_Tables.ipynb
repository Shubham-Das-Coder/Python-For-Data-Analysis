```python
import numpy as np
import pandas as pd
import os
```

```python
table1=pd.DataFrame({"PID":(1,2,3,4,5,6,7,8),
                    "gender":("male","male","female","female","female","male","female","male"),
                    "height":(71,73,64,64,66,69,62,72),
                    "weight":(175,225,130,125,165,160,115,250)})
table1
```

```python
table2=pd.DataFrame({"PID":(1,2,4,5,7,8,9,10),
                    "sex":("male","male","female","female","female","male","male","female"),
                    "visits":(1,2,4,12,2,2,1,1),
                    "checkup":(1,1,1,1,1,1,0,0),
                    "followup":(0,0,1,2,0,0,0,0),
                    "illness":(0,0,2,7,1,1,0,0),
                    "surgery":(0,0,0,2,0,0,0,0),
                    "ER":(0,1,0,0,0,0,1,1)})
table2
```

```python
combined1=pd.merge(table1,     #First table to be joined
                  table2,      #Second table to be joined
                  how="inner", #Merge method
                  on="PID")    #Columns to join
combined1
```

```python
left_join=pd.merge(table1,   #First table to be joined
                  table2,    #First table to be joined
                  how="left",#Merge method
                  on="PID")  #Columns to join
left_join
```

```python
right_join=pd.merge(table1,   #First table to be joined
                  table2,    #First table to be joined
                  how="right",#Merge method
                  on="PID")  #Columns to join
right_join
```

```python
outer_join=pd.merge(table1,   #First table to be joined
                  table2,    #First table to be joined
                  how="outer",#Merge method
                  on="PID")  #Columns to join
outer_join
```

```python
table1.rename(columns={"gender":"sex"},inplace=True)#Rename "gender column"
combined2=pd.merge(table1,         #First data frame            
                  table2,          #Second data frame
                  how="outer",     #Merge method
                  on=["PID","sex"]) #Columns to join
combined2
```
